```
Lütfen notlarını görmek istediğiniz öğrencinin adını girin: Mehmet

Mehmet isimli öğrencinin      1.Sınav Notu:72
                              2.Sınav Notu:66
                              3.Sınav Notu:48
Not Ortalaması: 62.0


````python
# 3.Problemin Çözümünü Buraya Yazınız
not_kisi = input("Lütfen notlarını görmek istediğiniz öğrencinin adını girin")
okul_kisi = {
    "Mehmet" : {"1. Sınav Notu" : 72, "2. Sınav Notu" : 66, "3. Sınav Notu" : 48},
    "Ahmet" : {"1. Sınav Notu" : 55, "2. Sınav Notu" : 35, "3. Sınav Notu" : 70},
    "Feyyaz" : {"1. Sınav Notu" : 39, "2. Sınav Notu" : 41, "3. Sınav Notu" : 90}
}
ortalama = okul_kisi[not_kisi]["1. Sınav Notu"] * okul_kisi[not_kisi]["2. Sınav Notu"] * okul_kisi[not_kisi]["3. Sınav Notu"] / 3
print(f"{not_kisi} isimli öğrencinin 1.Sınav Notu:{okul_kisi[not_kisi]["1. Sınav Notu"]}, 2.Sınav Notu:{okul_kisi[not_kisi]["2. Sınav Notu"]}, 3.Sınav Notu: {okul_kisi[not_kisi]["3. Sınav Notu"]}'dir. Not Ortalaması: {ortalama}dır.")
````
