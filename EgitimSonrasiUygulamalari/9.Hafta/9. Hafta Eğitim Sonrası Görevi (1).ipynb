# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import random
import time
class Kumanda():
    def __init__(self):
        self.tv_acik = False
        self.ses_ayarlari = 10
        self.kanal_ekle = []
        self.aktif_kanallar = None
    def tv_ac(self):
        if not self.tv_acik :
            self.tv_acik = True
            print("Televizyon açıldı")
        else: print("Televizyon zaten açık")
    
    def tv_kapa(self):
        if self.tv_acik:
            self.tv_acik = False
            print("Televizyon kapandı")
        else: print("Televizyon zaten kapalı")

    def ses_ayar(self):
        if self.tv_acik:
            print("Televizyon kapalı. İşlem sağlanamamaktadır.")
            return
        yeni_deger = input("0-100 arasında yeni ses seviyesini girin:")
        try:
            yeni_ses = int(yeni_deger)
            if 0 <= yeni_ses <= 100:
                self.ses_ayarlari = yeni_ses
                print(f"Ses seviyesi: {yeni_ses}")
            else: print("1-100 arasında bilgi giriniz")
        except ValueError:
            print("Geçersiz veri girildi.")
    
    def kanal_ekleme(self):
        if not self.tv_acik:
            print("TV kapalı. Kanak ekleme sağlanmaz")
            return
        kanal = input("Kanal ekle")
        if kanal not in self.kanal_ekleme():
            self.kanal_ekle.append(kanal)
            print(f"{kanal} kanalı eklendi")
        else: print("Bu kanal zaten var")
    
    def kanal_sayisi(self):
        if not self.tv_acik:
            print("Televizyon kapalı. Kanal sayısı görüntülenememektedir.")
            return
        else: print(f"Toplam kanal sayisi: {len(self.kanal_ekle)}")
    
    def rastgele_kanal(self):
        if not self.tv_acik:
            print("Televizyon kapalı")
            return
        if not self.kanal_ekle:
            print("Kanal listesi boş")
            return
        self.aktif_kanallar = random.choice(self.kanal_ekle)
        print(f"Rastgele seçilen kanal : {self.aktif_kanallar}")
    
    def televizyon_sonuc(self):
        if not self.tv_acik:
            print("Televizyon kapalı")
            return
        print(f"Televizyon Açık : {'Evet' if self.tv_acik else'Hayır'}")
        print(f"Ses seviyesi: {self.ses_ayarlari}")
        print(f"Mevcut Kanallar: {",".join(self.kanal_ekle) if self.kanal_ekle else 'Kanal yok'}")
        print(f"Aktif Kanallar: {self.aktif_kanallar if self.aktif_kanallar else 'Yok'}")
    
    def calistir(self):
        while True:
            if not self.tv_acik:
                print("Televizyon Kapalı")
                self.tv_acik()
            else:
                print("Seçim yapınız")
                print("1. TV Aç")
                print("2. TV Kapat")
                print("3. Ses Ayarları")
                print("4. Kanal Ekle")
                print("5. Kanal Sayısını Öğrenme")
                print("6. Rastgele Kanala Geçme")
                print("7. Televizyon Bilgileri")
                print("q. Çıkış")

                secim = input("Seçim yapınız").lower()

                if secim == "1":
                    self.tv_ac()
                elif secim == "2":
                    self.tv_kapa()
                elif secim == "3":
                    self.ses_ayar()
                elif secim == "4":
                    self.kanal_ekleme()
                elif secim == "5":
                    self.kanal_sayisi()
                elif secim == "6":
                    self.rastgele_kanal()
                elif secim == "q":
                    print("Çıkış sağlanıyor")
                    break
                else: print("Geçersiz seçim")

                time.sleep(1)

kumanda = Kumanda()
kumanda.calistir()
```
