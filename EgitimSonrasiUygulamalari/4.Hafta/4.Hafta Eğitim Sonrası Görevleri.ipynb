Aldƒ±ƒüƒ±nƒ±z √∂ƒürenci bilgilerini yine aynƒ± d√∂ng√º yardƒ±mƒ±yla `ogrenciler` adƒ±ndaki s√∂zl√ºkte saklayƒ±nƒ±z.
````python
# 1.Problemin √á√∂z√ºm√ºn√º Buraya Yazƒ±nƒ±z
ogrenciler = {}

for i in range(3):
    okul_no = input("√ñƒürencinin okul numarasƒ±nƒ± giriniz: ")
    isim = input("√ñƒürencinin adƒ±nƒ± ve soyadƒ±nƒ± giriniz: ")
    telefon = input("√∂ƒürencinin telefon numarasƒ±nƒ± giriniz: ")
    dogum_tarihi = input("√ñƒürencinin doƒüum tarihini giriniz : ")
    
    ogrenciler[okul_no] = (isim, telefon, dogum_tarihi)
print(ogrenciler)
````

### üå∂Ô∏è b) Hadi i≈üi biraz daha zorla≈ütƒ±ralƒ±m! 


≈ûimdiyse senden `numbers` isimli listedeki tekrar eden elemanlarƒ± `repeatNumbers` adƒ±ndaki listeye; tekrar etmeyen elemanlarƒ± `uniqueNumbers`adƒ±ndaki listeye kopyalayan yazmanƒ± ve listeleri ekrana bastƒ±rmanƒ± bekliyoruz

```
>Output
Tekrar Eden Elemanlar: [10, 10, 15, 15, 15, 20, 20, 20, 20]
Tekrar Etmeyen Elemanlar: [5]
```

````python
# 2.Problemin b ≈ûƒ±kkƒ±nƒ±n √á√∂z√ºm√ºn√º Buraya Yazƒ±nƒ±z
numbers = [5,10,10,15,15,15,20,20,20,20]
repeatNumbers = []
uniqueNumbers = []
for i in numbers : 
    if numbers.count(i) > 1 : 
        repeatNumbers.append(i)
    else : uniqueNumbers.append(i)
print("Tekrar Eden Elemanlar: ", repeatNumbers)
print ("Tekrar Etmeyen Elemanlar: ", uniqueNumbers)
````
