# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız
def main():
    kisi = int(input("Kisi sayısı:"))
    for i in range(kisi):
        isim = input("isim")
        yas = input("Yas")
        kilo = input("Kilo")
        boy = input("Boy")
        

        vki = vki_hesap(kilo, boy)
        deger_vki = deger(vki)
        cikt_str = cikti(isim, yas, kilo, boy, vki, deger_vki)

        with open("vki.text", "a") as file:
            file.write(cikt_str)
            file.write("*"*20)
            print("Kayıt edildi!")
    print("Görev tamamlandı")

       

def vki_hesap(kilo, boy):
        vki = int(kilo) / (int(boy)**2)
        return round(vki, 1)
    
def deger(vki):
    if vki < 18.5:
        return "Zayıf"
    elif vki < 24.9 : 
        return "Normal"
    elif vki < 29.9:
        return "Fazla"
    elif vki < 39.9 :
        return "Obez"
    else: return "İleri derecede obez"

def cikti(isim,yas, boy, kilo, vki, deger):
    c1 = f"İsim - Soyisim : {isim}\n"
    c2 = f"Yas : {yas}, Boy : {boy}, Kilo : {kilo}\n"
    c3 = f"Vücut Kitle İndeksi {vki}\n"
    c4 = f"Değer aralığı {deger}\n"
    return c1 + c2 + c3 + c4

main()
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız
import datetime
menu = { 
    "Yeşil Smoothie": 15.0,
    "Mango Smoothie": 20.0,
    "Yaban Mersinli Smoothie": 18.0,
    "Protein Shake": 22.0
}

satin_alinanlar = []

def main():
    print("1.Menüyü Görüntüle")
    print("2. Satın Alma")
    print("3. Satın Alınanlar")
    print("4. Çıkış")

    secim = input("Seçim Yap")

    if secim  == "1":
        menu_goster()
    elif secim == "2":
        sec_urun= input("Ürün adlarını giriniz.").split(",")
        sec_urun = [urun.strip() for urun in sec_urun]
        satinAlim(sec_urun)
    elif secim == "3":
        kayit()
    elif secim == "4":
        print("Çıkış sağlanıyor.")
    else: print("Geçersiz")

def menu_goster():
    print("Menü")
    for urun, ucret in menu.items():
        print(f"{urun}: ${ucret:.2f}\n")
    
def toplam_fiyat(secilen):
    toplam = 0
    for urun in secilen:
        if urun in menu:
                toplam += menu[urun]
        else:
            print(f"{urun} menüde mevcut değil.")
    return toplam

def satinAlim(secilen):
    toplam = toplam_fiyat(secilen)
    if toplam > 0:
        now = datetime.datetime.now()
        satin_alinanlar.append({
            "urun": secilen,
            "ucret": toplam,
            "tarih": now.strftime("%Y-%m-%d %H:%M:%S")
        })
        print(f"Toplam Tutar: ${toplam:.2f}")
    else: print("Satın alma işlemi yapılamadı.")

def kayit():
    if satin_alinanlar:
        for i in  satin_alinanlar:
            items = ",".join(i["urun"])
            print(f"Tarih: {i["date"]}, Ürünler: {items}, Toplam: ${i["ucret"]:.2f}")
    else: print("Satın alım sağlanmadı")

main()
```
